```python
import  cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('cat.jpg')
image2 = cv2.imread('ronaldo.jpg')
image3 = cv2.imread('messi.jpg')
image4 = np.zeros((512,512,3),np.uint8)
```

```python
b,g,r = cv2.split(image1)
cv2.imshow('Blue',b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged_image = cv2.merge((b,g,r))
cv2.imshow('Merged image',merged_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
croped_image = image1[0:200,250:500]
cv2.imshow('Croped image',croped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image1 = cv2.resize(image2,(512,512))
resized_image2 = cv2.resize(image3,(512,512))

cv2.imshow('Resized image',resized_image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
added_image = cv2.add(resized_image1,resized_image2)

added_image = cv2.addWeighted(resized_image1,.2,resized_image2,0.8,0) #0 for brightness
cv2.imshow('Merged 2 images',added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flipped_image = cv2.flip(image2, 2)
cv2.imshow('Flipped image',flipped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
