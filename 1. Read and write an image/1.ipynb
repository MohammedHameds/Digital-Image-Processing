```python
import cv2
```

```python
print(cv2.__version__)
```

```python
image = cv2.imread('fly.jpg' , 1)
cv2.imshow('My Image' , image)   #BGR
k = cv2.waitKey(0)
if k == 27 :
    cv2.destroyAllWindows()
elif k == ord('s') :
    cv2.imwrite('New Image.png' , image)
    cv2.destroyAllWindows()    
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)   #RGB
plt.xticks([])
plt.yticks([])
plt.show()      
```
