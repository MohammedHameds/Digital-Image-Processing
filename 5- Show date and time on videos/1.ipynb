```python
import cv2
import datetime
```

```python
cap = cv2.VideoCapture(0)
cap.set(3,1080)
cap.set(4,720)

while (cap.isOpened()):
    ret,frame = cap.read()
    
    if ret == True :
        
        height = int(cap.get(4))
        width = int(cap.get(3))
        time = str(datetime.datetime.now())

        text = f'Height :{height} , width :{width}'   
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, text, (20,40) , font , 1 , (255,0,255) , 2) 
        cv2.putText(frame, time, (20,80) , font , 1 , (0,255,255) , 2) 
        

        cv2.imshow('My video' , frame)
        if cv2.waitKey(1) == ord('q') :
            break
    else :
        break
    
cap.release()    
cv2.destroyAllWindows()
```
