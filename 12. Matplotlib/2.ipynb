```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('cat.jpg',0)
```

```python
ret,th1 = cv2.threshold(image,150,255,cv2.THRESH_BINARY)
ret,th2 = cv2.threshold(image,150,255,cv2.THRESH_BINARY_INV)
ret,th3 = cv2.threshold(image,150,255,cv2.THRESH_TRUNC)
ret,th4 = cv2.threshold(image,150,255,cv2.THRESH_TOZERO)
ret,th5 = cv2.threshold(image,150,255,cv2.THRESH_TOZERO_INV)
th6 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY,111,2)
th7 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY,111,2)
```

```python
names = ['Original','Binary','Binary inverse','Trunc','Tozero',
         'Tozero inverse','Adaptive gaussian' , 'Adaptive mean']

filters  = [image,th1,th2,th3,th4,th5,th6,th7]

plt.figure(figsize=(20, 7))

for i in range(8) :
    plt.subplot(2, 4, i+1)
    img = cv2.cvtColor(filters[i],cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(names[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
