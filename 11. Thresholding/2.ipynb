```python
import cv2
```

```python
image = cv2.imread('cat.jpg',0)
```

```python
ret,th1 = cv2.threshold(image,150,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY,111,2)
th3 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY,111,2)


cv2.imshow('Original image',image)
cv2.imshow('Binary threshold image',th1)
cv2.imshow('Adaptive Threshold gaussian image',th2)
cv2.imshow('Adaptive Threshold mean image',th3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
