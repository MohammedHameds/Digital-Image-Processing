```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pic1.jpg',1)
Laplacian = cv2.Laplacian(image,-1,ksize = 3)
Lap = np.absolute((Laplacian))
```

```python
names = ['Original','Laplacian','Lap']
filters  = [image,Laplacian,Lap]

plt.figure(figsize=(15, 7))
for i in range(3) :
    plt.subplot(1, 3, i+1)
    img = cv2.cvtColor(filters[i],cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(names[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
