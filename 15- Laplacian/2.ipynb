```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sudoco.png',1)
```

```python
sobel_x = cv2.Sobel(image,cv2.CV_64F,1,0)
sobel_y = cv2.Sobel(image,cv2.CV_64F,0,1)
sobel_x = np.uint8(np.absolute((sobel_x)))
sobel_y = np.uint8(np.absolute((sobel_y)))
bitwise_or = cv2.bitwise_or(sobel_x,sobel_y)
```

```python
names = ['Original','Sobel_x','Sobel_y','Bitwise_or']
filters  = [image,sobel_x,sobel_y,bitwise_or]

plt.figure(figsize=(12, 10))
for i in range(len(names)) :
    plt.subplot(2, 2, i+1)
    img = cv2.cvtColor(filters[i],cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(names[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
