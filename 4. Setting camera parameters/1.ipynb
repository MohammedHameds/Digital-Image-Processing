```python
import cv2
```

```python
cap = cv2.VideoCapture(0)

cap.set(3,3000)
cap.set(4,3000)

while (cap.isOpened()) :
    ret,frame = cap.read()
    
    if ret == True :
        cv2.imshow('My video',frame)
        
        if cv2.waitKey(27) == ord('q') :
            break
    else : 
        break
cap.release()
cv2.destroyAllWindows()               
```
