```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('Gray Video.avi',fourcc,20,(640, 480))
out_HSV = cv2.VideoWriter('HSV Video.avi',fourcc,20,(640, 480))

while(cap.isOpened()): 
 ret,frame = cap.read()


 if ret == True :
     
     _HSV = cv2.cvtColor(frame,cv2.COLOR_RGB2HSV)
         
     out.write(frame)
     out_HSV.write(_HSV)
     
     cv2.imshow('Gray',frame)
     cv2.imshow('HSV',_HSV)
     
     
     if cv2.waitKey(27) == ord('q') :
         break
 else:
     break
        
cap.release()
out.release()
out_HSV.release()
cv2.destroyAllWindows()        
```

```python
cap = cv2.VideoCapture('Gray Video.avi')


while (cap.isOpened() == True ) :
    ret , fram = cap.read()
    if ret == True:
     gray = cv2.cvtColor(fram, cv2.COLOR_BGR2GRAY)
    
     cv2.imshow('Frams' , gray)

     if cv2.waitKey(27)  == ord ('q') : 
         break
    else : 
         break
    

cap.release()
cv2.destroyAllWindows()      
                
```
