```python
import cv2
import numpy as np
```

```python
image = cv2.imread('sudoco.png',0)
```

```python
lr1 = cv2.pyrDown(image)
lr2 = cv2.pyrDown(lr1)

hr1 = cv2.pyrUp(lr2)
hr2 = cv2.pyrUp(hr1)
```

```python
cv2.imshow('Original',image)
cv2.imshow('hr1',hr1)
cv2.imshow('hr2',hr2)
cv2.imshow('lr1',lr1)
cv2.imshow('lr2',lr2)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
