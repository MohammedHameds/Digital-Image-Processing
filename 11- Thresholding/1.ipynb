```python
import cv2
```

```python
image = cv2.imread('cat.jpg')
```

```python
ret,tr1 = cv2.threshold(image,150,255,cv2.THRESH_BINARY)
ret,tr2 = cv2.threshold(image,150,255,cv2.THRESH_BINARY_INV)
ret,tr3 = cv2.threshold(image,150,255,cv2.THRESH_TRUNC)
ret,tr4 = cv2.threshold(image,150,255,cv2.THRESH_TOZERO)
ret,tr5 = cv2.threshold(image,150,255,cv2.THRESH_TOZERO_INV)

cv2.imshow('Original image',image)
cv2.imshow('Binary threshold image',tr1)
cv2.imshow('Binary threshold inverse image',tr2)
cv2.imshow('Trunc threshold image',tr3)
cv2.imshow('Tozero threshold image',tr4)
cv2.imshow('Tozero inverse threshold image',tr5)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
