```python
import cv2
import numpy as np
```

```python
image = cv2.imread('ball.png')
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)


low = np.array([50,10,50]) 
upper = np.array([130,250,250])
mask = cv2.inRange(image, low, upper)
bi_And = cv2.bitwise_and(image, image, mask = mask)    

cv2.imshow('image',image)
cv2.imshow('hsv',hsv)
cv2.imshow('mask',mask)
cv2.imshow('bi_And',bi_And)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
